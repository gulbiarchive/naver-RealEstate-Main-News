```python
import requests
import time
import pandas as pd

datas = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

for i in range(1, 20):
    url = f"https://land.naver.com/news/airsList.naver?baseDate=2024-08-05&page={i}&size=20"
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # HTTP 에러가 발생하면 예외를 일으킵니다.
        
        data = response.json()
        if 'list' not in data or not data['list']:  # JSON 데이터가 비어 있다면 반복문을 빠져나갑니다.
            break
        
        datas.append(data)
    
    except requests.exceptions.RequestException as e:
        print(f"{i} 페이지에서 데이터를 가져오지 못했습니다. 에러: {e}")
        break
    
    time.sleep(1)  # 요청 간에 1초 간격을 둡니다.

land = []

for page_data in datas:
    for article in page_data.get('list', []):
        date = article.get('publishDateTime', '')[:10]
        title = article.get('title', '')
        link = article.get('linkUrl', '')
        summary = article.get('summaryContent', '').split('.')[0] + '.'

        land.append([date, title, link, summary])

df = pd.DataFrame(land, columns=['날짜', '제목', '링크', '요약'])
df.index += 1
df
```
